```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
import json
```

```python
def plot_rank(n):
    with open(f"rank_{n}.json", "r") as f:
        d = json.load(f)
    
    grid = np.array(d['data'])
    X = np.zeros_like(grid)
    Y = np.zeros_like(grid)
    s = grid.shape
    for i in range(d['nx']):
        for j in range(d['ny']):
            X[j, i] = d['x0'] + i * d['dx']
            Y[j, i] = d['y0'] + j * d['dy']

    #plt.imshow(Y, vmin = -1, vmax = 1)
    plt.pcolormesh(X, Y, grid, vmin = -1, vmax = 1)
```

```python
plot_rank(4)
```

```python
for i in range(16):
    plot_rank(i)
```
